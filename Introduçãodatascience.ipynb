```python
#fazendo uma pesquuisa em seu dataframe
# estou solicitando filmes com nota = 1 e solicitei a media notas   do filme cujo ID = 1,nesse caso TOY STORE

notas.query('filmeId==1').nota.mean()
```

```python
notas.query('filmeId==2').nota.mean()
```

```python
# agora eu quero a media da media por filme, AGRUPAR
# o comando 'notas.groupby ('filmeId')' agrupou todos os filmes, e agora o que podemos fazer com isso, tiramos a media, mas Ã© qualque media ?NAO, quero a media das notas por isso comando 'nota.mean()'
# mas tem uma coisa que Ã© importante, Ã© as boas praticas em programaÃ§Ã£o por isso iremos criar outra VARIAVEIS para ficar claro o que estamos fazendo
medias_por_filme = notas.groupby ('filmeId').nota.mean()
medias_por_filme.head()
```

```python
medias_por_filme.plot(kind = 'hist')
```

```python
sns.boxplot(medias_por_filme)
```

```python
medias_por_filme.describe()
```
