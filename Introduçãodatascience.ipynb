```python
# Explorando os dados, a ideia agora é explorar melhor esses dados. Até aqui vimos algumas informações sobre as notas media, mediana e grafico. Qual coluna podemos explocar mais ?

filmes = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/ml-latest-small/movies.csv")
filmes.head()
```

```python
filmes.columns = ['filmeId', 'titulo', 'generos']
filmes.head()
```

```python
notas.head()
```

```python
#fazendo uma pesquuisa em seu dataframe
# estou solicitando filmes com nota = 1 e solicitei a media notas   do filme cujo ID = 1,nesse caso TOY STORE

notas.query('filmeId==1').nota.mean()
```

```python
notas.query('filmeId==2').nota.mean()
```

```python
# agora eu quero a media da media por filme, AGRUPAR
# o comando 'notas.groupby ('filmeId')' agrupou todos os filmes, e agora o que podemos fazer com isso, tiramos a media, mas é qualque media ?NAO, quero a media das notas por isso comando 'nota.mean()'
# mas tem uma coisa que é importante, é as boas praticas em programação por isso iremos criar outra VARIAVEIS para ficar claro o que estamos fazendo
medias_por_filme = notas.groupby ('filmeId').nota.mean()
medias_por_filme.head()
```

```python
medias_por_filme.plot(kind = 'hist')
```

```python
sns.boxplot(medias_por_filme)
```

```python
medias_por_filme.describe()
```

```python
# nos podemos personalizar mais ainda o grafico podemos baixar mais uma biblioteca para auxiliar nesse processo

import matplotlib.pyplot as plt
```

```python
#resumo com matplotlib podemos personalizar o grafico com caso acima colocamos nome no grafico;
sns.distplot(medias_por_filme)
plt.title('Histograma das médias dos filmes')
```

```python
tmdb = pd.read_csv('/content/tmdb_5000_movies.csv')
tmdb.head()
```

```python
#aqui ficou diferente porque tive que colocar colchetes e parenteses para chamar o revenue que é faturamento
#plt.shoe retira dados, queremos somente o grafico
sns.displot(tmdb['revenue'])
plt.title('Distribuição das Receitas dos filmes')
plt.show()
```

```python
# prompt: grafico de distribuicao do orçamento dos filmes (budget)
# Criando o histograma da distribuição do orçamento (budget)
sns.histplot(tmdb['budget'], kde=True) # kde=True adiciona a curva de densidade
plt.title('Distribuição do Orçamento dos Filmes')
plt.xlabel('Orçamento')
plt.ylabel('Número de Filmes')
plt.show()
```
